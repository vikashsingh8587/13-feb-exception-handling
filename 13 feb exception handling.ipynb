```python
# Q4. Why LookupError class is used? Explain with an example KeyError and IndexError.
```

```python
# lookup error
#LookupError : this is the base exception for those are raised when a key and index used on\
#mapping or squence is invalid or not found . the exception are raised are:
#KeyError
#IndexError
#IndexError
import logging
logging.basicConfig(filename="Exception.log",level = logging.INFO)
try :
    l=[1,22,0]
    logging.info(l[4])
except IndexError:
    logging.info("this is raise index error")
logging.shutdown()    
# key error
import logging
logging.basicConfig(filename="Exception.log",level = logging.INFO)
try :
    d={"a":1,"b":2}
    logging.info(d["b"])
except KeyError:
    logging.info("this is raise key error")
logging.shutdown() 
```

```python
# Q6. List down some best practices for exception handling in python.
```

```python
# use always of specific specfic exception 
try:
    10/0
except Exception as a:
        print(a)
```

```python
import logging
logging.basicConfig(filename ="error.log",level = logging.ERROR)
try:
    10/0
except ZeroDivisionError as e:
    logging.error("i am try to handle a zerodivsion error {}".format(e))
```

```python
# always avoid the write the multiple exception
```

```python
try:
    10/0
except FileNotFoundError as e:
    logging.error("i am handling file not found {}".format(e))
except AttributeError as e:
    logging.error(" i am handling atrribute error{}".format(e))
except ZeroDivisionError as e:
    logging.error("i am handling Zerodivision error{}".format(e))
```
